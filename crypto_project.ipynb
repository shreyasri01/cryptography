```python
from tkinter import *
from tkinter import messagebox
import base64
import os


def decrypt():
    a="Dem01234"
    password=code.get()
    
    if password==a:
        screen2=Toplevel(screen)
        screen2.title("decryption")
        screen2.geometry("400x200")
        screen2.configure(bg="light blue")
        
        message=text1.get(1.0,END)
        decode_message=message.encode("ascii")
        base64_bytes=base64.b64decode(decode_message)
        decrypt=base64_bytes.decode("ascii")
        
        
        Label(screen2,text="DECRYPT",font="arial",fg="white",bg="#00bd56").place(x=10,y=0)
        text2=Text(screen2,font="arial", bg="white",relief=GROOVE,wrap=WORD,bd=0)
        text2.place(x=10,y=40,width=300,height=150)
        
        text2.insert(END,decrypt)
        
    elif password=="":
        messagebox.showerror("encryption","Input Password")
        
    elif password!=a:
            messagebox.showerror("encryption","Invalid Password")
        
        
    
def encrypt():
    a="Dem01234"
    password=code.get()
    
    if password==a:
        screen1=Toplevel(screen)
        screen1.title("encryption")
        screen1.geometry("400x200")
        screen1.configure(bg="wheat")
        
        message=text1.get(1.0,END)
        encode_message=message.encode("ascii")
        base64_bytes=base64.b64encode(encode_message)
        encrypt=base64_bytes.decode("ascii")
        
        
        Label(screen1,text="ENCRYPT",font="arial",fg="white",bg="#ed3833").place(x=10,y=0)
        text2=Text(screen1,font="arial", bg="white",relief=GROOVE,wrap=WORD,bd=0)
        text2.place(x=10,y=40,width=300,height=150)
        
        text2.insert(END,encrypt)
        
    elif password==" ":
        messagebox.showerror("encryption","Input Password")
        
    elif password!=a:
        messagebox.showerror("encryption","Invalid Password")
        

def main_screen():
    global screen
    global code 
    global text1
    
    screen = Tk()
    screen.geometry("400x300")
    
    #icon
    screen.title("CodeWords")
        
        
        


    def reset():
        code.set("")
        text1.delete(1.0,END)
    
    Label(text="Enter Text For Encryption and Decryption", fg="purple",font=("calbri",13)).place(x=12,y=14)
    text1 = Text(font="Roboto 20", bg="pink",relief=GROOVE,wrap=WORD,bd=0)
    text1.place(x=10,y=60,width=350,height=200)
    Label(text="Enter secret key for encryption and decryption",fg="purple",font=("calbri",13)).place(x=12,y=180)
    
    code=StringVar()
    Entry(textvariable=code,width=15,bd=0,font=("arial",30)).place(x=17,y=210)
    
    Button(text="ENCRYPT",height="2",width=24,bg="Black",fg="pink",bd=0,command=encrypt).place(x=10,y=270)
    Button(text="DECRYPT",height="2",width=24,bg="Black",fg="pink",bd=0,command=decrypt).place(x=190,y=270)
    Button(text="RESET",height="2",width=48,bg="Black",fg="white",bd=0,command=reset).place(x=10,y=320)
    
    screen.mainloop()
    
main_screen()
```
